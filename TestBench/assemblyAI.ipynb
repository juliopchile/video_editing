```python
# OUTPUT

# 2. Export SRT subtitles (chars_per_caption configurable)
srt_subtitles = transcript.export_subtitles_srt(chars_per_caption=32)
srt_filename = f"transcript_{transcript.id}.srt"
with open(srt_filename, "w", encoding="utf-8") as srt_file:
    srt_file.write(srt_subtitles)
print(f"\nSRT subtitles saved as: {srt_filename}")

# (Optional) Export VTT subtitles
vtt = transcript.export_subtitles_vtt()
vtt_filename = f"transcript_{transcript.id}.vtt"
with open(vtt_filename, "w", encoding="utf-8") as vtt_file:
  vtt_file.write(vtt)

# 3. Print paragraph segmentation (if available)
paragraphs = transcript.get_paragraphs()
if paragraphs:
    print("\n=== Paragraphs ===")
    for para in paragraphs:
        print(para.text)
        print()

words = transcript.words  # list of Word objects
words_filename = f"transcript_{transcript.id}_words.txt"

with open(words_filename, "w", encoding="utf-8") as wf:
    # Header (optional)
    wf.write("word\tstart_ms\tend_ms\tspeaker\n")
    for w in words:
        # w.text: the word string
        # w.start: start time in ms
        # w.end:   end time in ms
        # w.speaker_label: only if speaker_labels=True
        speaker = getattr(w, "speaker_label", "n/a")
        wf.write(f"{w.text}\t{w.start}\t{w.end}\t{speaker}\n")

# 4. (Example) Keyword search in the transcript
keywords = ["lean", "Eminem", "Freestyle"]
matches = transcript.word_search(keywords)
if matches:
    print("=== Keyword Search Results ===")
    for match in matches:
        print(f"Found '{match.text}' {match.count} times")
else:
    print("No keyword matches found.")

```

```python
import os
import re

def limpiar_nombre_archivo_con_path(path_completo):
    # 1. Separar directorio y nombre de archivo
    directorio, filename = os.path.split(path_completo)
    # 2. Separar nombre y extensión
    nombre, extension = os.path.splitext(filename)
    # 3. Quitar número inicial seguido de guión bajo
    nombre = re.sub(r'^\d+_', '', nombre)
    # 4. Quitar sufijo _(Vocals) (case‑insensitive)
    nombre = re.sub(r'_\(vocals\)$', '', nombre, flags=re.IGNORECASE)
    # 5. Reconstruir nombre limpio con extensión
    archivo_limpio = nombre + extension
    # 6. Volver a juntar ruta
    return os.path.join(directorio, archivo_limpio)

# — Ejemplos —
archivos = [
    r"bin\clean_audio\1_Marithea_vs_ElMenor_FMS_Int_NoSubs_(Vocals).wav",
    r"bin\clean_audio\1_MINUTAZO de BNET - MÉTRICAS ANALIZADAS - BNET VS TIRPA FMS 2020_(Vocals).wav",
    r"bin\clean_audio\Marithea_vs_ElMenor.wav",
    r"bin\clean_audio\2_Pista_(VOCALS).mp3",
]

for path in archivos:
    print("Original:", path)
    print("Cleaned: ", limpiar_nombre_archivo_con_path(path))
    print()
```
